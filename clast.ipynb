```python
df_1 = pd.read_csv('data/client_data_apdated.csv')
df_2 = pd.read_csv('data/client_data_clusters.csv')

# Если индексы строк совпадают
result = pd.concat([df_1[0:600], df_2], axis=1)
```

```python
result
```

```python
# Создание профилей кластеров
cluster_summary = result.groupby('cluster_KMeans').agg({
    'age': 'mean',
    'income': 'mean',
    'personal_loan': 'mean',
    'creditcard': 'mean',
    'loan_delinquency': 'mean',
    'family': 'mean'
}).round(2)

# Добавление размера кластера
cluster_summary['Size'] = result['cluster_KMeans'].value_counts().sort_index()
cluster_summary['Percentage'] = (cluster_summary['Size'] / len(result) * 100).round(2)

print("\nДетальный профиль кластеров:")
print(cluster_summary)

# Описание кластеров (пример)
cluster_descriptions = {
    0: "Молодые клиенты с низким доходом, редко используют кредитные продукты",
    1: "Клиенты предпенсионного возраста со стабильным доходом",
    2: "Состоятельные клиенты среднего возраста, активно пользующиеся кредитами",
    3: "Высокорисковые клиенты с просрочками"
}

d = {}
for cluster_num, description in cluster_descriptions.items():
    if cluster_num in result['cluster_KMeans'].values:
        size = len(result[result['cluster_KMeans'] == cluster_num])
        d[cluster_num] = ['Описание', size, ]
        print(f"\nКластер {cluster_num} ({size} клиентов, {size/len(result)*100:.1f}%):")
        print(f"Описание: {description}")
```
